##### 主要な変数の説明
- `--source`: WSIのsvsファイルが置かれているディレクトリ
- `--save_dir`: CLAMの出力を保存するディレクトリ
- `--patch_size`: パッチの大きさ
- `--step_size`: デフォルトが256なので注意。パッチの大きさと等しくすればoverlapなし
- `--seg`: 組織領域のsegmentationを行う
- `--patch`: パッチを切り出す
- `--stitch`: 抽出したパッチを基にWSIを再構成する
